## Let's compare the implementation of neural network by for loop and vectorization 

```python
#Sigmoid function
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# for loops
x = np.array([200, 17])
W = np.array([[1, -3, 5],
              [-2, 4, 6]])
b = np.array([-1, 1, 2])

def dense(a_in, W, b):
    units = W.shape[1]
    a_out = np.zeros(units)
    for i in range(units):
        w = W[:, i]
        z = np.dot(a_in, w) + b[i]
        a_out[i] = sigmoid(z)
    return a_out    

#execution
output = dense(x, W, b)
print(output)

```

```python
# vectorization
X = np.array([[200, 17]])
W = np.array([[1, -3, 5],
              [-2, 4, 6]])
B = np.array([[-1, 1, 2]])

def dense(A_IN, W, B):
    Z = np.matmul(A_IN, W) + B
    A_OUT = sigmoid(Z)
    return A_OUT
#execution 
output = dense(X, W, B)
print(output)
```
